```python
# Remove estados duplicados
aux_list = set()
unique_list_o = []
for item in estados_teste_o:
    aux =  tuple([element for sublist in item for element in sublist])
    if aux not in aux_list:
        unique_list_o.append(item)
        aux_list.add(aux)
array_of_o = ['O'] * len(unique_list_o)
```

```python
unique_list = unique_list_x + unique_list_o
array_players = array_of_x + array_of_o
```

```python
def count_none_elements(array_of_arrays):
    count = 0
    for subarray in array_of_arrays:
        count += subarray.count(None)
    return count



initial_state_1 = [['O',  None, 'X'],
                [ None,  'O', 'X'],
                [ None, None, None]]

initial_state_2 = [[None,  None, None],
                [ None,  'O', 'X'],
                [ None, None, None]]

sel_algorithm = ['pruning', 'minmax']

dict = {
    'empt_spaces': [],
    'poda': [],
    'start_player': [] ,
    'time_sec': [],
    'nos_creat': []
}
for algo in sel_algorithm:
    for index, order_payer in enumerate(array_players):
        board = unique_list[index]
        game_over=False
        total_time = 0
        sel_algorithm = algo
        minMaxTicTacToe.initial_player = order_payer
        minMaxTicTacToe.states_visited = 0
        while game_over == False:

            game_over = minMaxTicTacToe.terminal(board)
            player = minMaxTicTacToe.player(board)
        
            if(game_over == False):
                #move = minMaxTicTacToe.minimax(board)
                if(sel_algorithm == 'pruning'):
                    tic = time.time()
                    move = minMaxTicTacToe.minimax(board)
                    toc = time.time()
                    time_taken = toc - tic
                    total_time = time_taken + total_time
                elif sel_algorithm == 'minmax':
                    tic = time.time()
                    move = minMaxTicTacToe.minimax_nopruning(board)
                    toc = time.time()
                    time_taken = toc - tic
                    total_time = time_taken + total_time
                if(move != None):
                    board = minMaxTicTacToe.result(board, move)
                            
                #print_board(board)    
                #print('#####################')
                #print('#####################')
                if(player == minMaxTicTacToe.O):
                    user = minMaxTicTacToe.X
                else:
                    user = minMaxTicTacToe.O
            else:
                #print('Algoritmo: ', algo)
                #print('Nós gerados: ', minMaxTicTacToe.states_visited)
                #print('Tempo total de execução: ', total_time)
                dict['empt_spaces'].append(count_none_elements(unique_list[index]))
                if(sel_algorithm == 'pruning'):
                    dict['poda'].append('SIM') 
                elif sel_algorithm == 'minmax':
                    dict['poda'].append('NAO')     
                dict['start_player'].append(order_payer)    
                dict['time_sec'].append(total_time)
                dict['nos_creat'].append(minMaxTicTacToe.states_visited)
                 
                                           

                
```
